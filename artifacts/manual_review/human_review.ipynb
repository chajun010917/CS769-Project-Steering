```python
import json

with open('artifacts/triples/triples_1000_sample_10232025.jsonl', 'r', encoding='utf-8') as f:
    data = [json.loads(line) for line in f]
```

```python
for item in data:
    wrong_chain = item['wrong_chain'].split('\n\n')
    item['wrong_chain_readable'] = wrong_chain
    item['bad_sample'] = None
```

```python
import json
with open('10232025_human_review.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

for datam in data:
    if 'answer' not in datam['wrong_chain_readable'][-1]:
        datam['bad_sample'] = True
    if f"answer is {datam['correct_answer']}" in datam['wrong_chain_readable'][-1]:
        datam['bad_sample'] = True
    elif f"answer is {datam['wrong_answer']}" in datam['wrong_chain_readable'][-1] and (datam['wrong_answer'] != datam['correct_answer']):
        datam['bad_sample'] = False
    if f"answer is {datam['correct_answer']}" in datam['wrong_chain_readable'][-1]:
        datam['bad_sample'] = True
    if f"answer: {datam['correct_answer']}" in datam['wrong_chain_readable'][-1]:
        datam['bad_sample'] = True

with open('10232025_human_review.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=2)
```
